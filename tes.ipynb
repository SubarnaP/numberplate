```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = "dataset/images/2024-03-08_18_5.jpg"
img = cv2.imread(image_path)
h, w = img.shape[:2]
```

```python
# YOLO label: class_id x_center y_center width height (normalized)
label_line = "0 0.3174829170352123 0.4931857789335804 0.368426682519131 0.17011612511332125"
_, x_center, y_center, bbox_width, bbox_height = map(float, label_line.strip().split())

# Convert normalized coordinates to absolute pixel values
x_center *= w
y_center *= h
bbox_width *= w
bbox_height *= h

x1 = int(x_center - bbox_width / 2)
y1 = int(y_center - bbox_height / 2)
x2 = int(x_center + bbox_width / 2)
y2 = int(y_center + bbox_height / 2)
```

```python
# Draw the bounding box on a copy of the image
img_with_box = img.copy()
cv2.rectangle(img_with_box, (x1, y1), (x2, y2), color=(0, 255, 0), thickness=2)

# Convert BGR to RGB for display
img_rgb = cv2.cvtColor(img_with_box, cv2.COLOR_BGR2RGB)

# Display using matplotlib
plt.figure(figsize=(8, 6))
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Image with Bounding Box")
plt.show()
```
